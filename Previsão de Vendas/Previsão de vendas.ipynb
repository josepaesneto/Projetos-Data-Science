#### Importar a Base de dados

```python
import pandas as pd

tabela = pd.read_csv("advertising.csv")
display(tabela)
print(tabela.info())
```

#### Análise Exploratória
- irei tentar visualizar como as informações de cada item estão distribuídas
- irei ver a correlação entre cada um dos itens

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(tabela)
plt.show()

sns.heatmap(tabela.corr(), annot=True, cmap="Wistia")
plt.show()
```

#### Com isso, posso partir para a preparação dos dados para treinar o Modelo de Machine Learning

- Separando em dados de treino e dados de teste

```python
from sklearn.model_selection import train_test_split

y = tabela["Vendas"]
x = tabela.drop("Vendas", axis=1)

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.6)
```

#### Possuo um problema de regressão - Irei escolher os modelos que vou usar:

- Regressão Linear
- RandomForest (Árvore de Decisão)
- RANSAC

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import RANSACRegressor

# cria as inteligencias artificiais
modelo_regressaolinear = LinearRegression()
modelo_arvoredecisao = RandomForestRegressor()
modelo_Ransac = RANSACRegressor()

#treinar as inteligencias artificiais

modelo_regressaolinear.fit(x_treino, y_treino)
modelo_arvoredecisao.fit(x_treino, y_treino)
modelo_Ransac.fit(x_treino, y_treino)
```

#### Teste da AI e Avaliação do Melhor Modelo

- Vou usar o R² -> diz o % que o modelo consegue explicar o que acontece

```python
from sklearn import metrics

# criar as previsoes
previsao_regressaolinear = modelo_regressaolinear.predict(x_teste)
previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)
previsao_Ransac = modelo_Ransac.predict(x_teste)

# comparar os modelos
print(metrics.r2_score(y_teste, previsao_regressaolinear))
print(metrics.r2_score(y_teste, previsao_arvoredecisao))
print(metrics.r2_score(y_teste, previsao_Ransac))
```

#### Visualização Gráfica das Previsões

```python
tabela_auxiliar = pd.DataFrame()
tabela_auxiliar["y_teste"] = y_teste
tabela_auxiliar["Previsoes ArvoreDecisao"] = previsao_arvoredecisao
tabela_auxiliar["Previsoes Regressao Linear"] = previsao_regressaolinear
tabela_auxiliar["Previsoes Ransac"] = previsao_Ransac
plt.figure(figsize=(15,6))
sns.lineplot(data=tabela_auxiliar)
plt.show()
```

#### Qual a importância de cada variável para as vendas?

```python
sns.barplot(x=x_treino.columns, y=modelo_arvoredecisao.feature_importances_)
plt.show()

# Caso queira comparar Radio com Jornal
#print(tabela[["Radio", "Jornal"]].sum())
```
